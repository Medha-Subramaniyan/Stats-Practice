```python
planes[planes['Price'].isna()] # there are 616 null rows! 
```

```python
planes = planes[~planes['Price'].isna()] 
```

```python
samp_price = planes['Price'].sample(n=30)
```

```python
pop_mean = planes['Price'].mean() #populaton mean 9044
samp_mean = samp_price.mean() # sample mean 10640.4
```

```python
abs_diff = abs(pop_mean - samp_mean)
abs_diff 
```

```python
sampMeans = []
for i in range(5):
    samp_test = planes['Price'].sample(n=30, replace=True)
    mean = samp_test.mean()
    sampMeans.append(mean)
```

```python
sampMeans
```

```python
for i in range(len(sampMeans)):
    print('Absolute difference btwn sample and pop mean', abs(sampMeans[i] - pop_mean))
```
